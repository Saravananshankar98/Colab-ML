```python
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,3,4,5]
y=[23,34,11,25,9]

plt.plot(x,y)
plt.show()
```

```python
## Object Oriented Approach
fig, ax = plt.subplots(2)
ax[0].plot(x,y)
ax[1].plot(x,y)
ax[0].set_title('first plot')
ax[1].set_title('second plot')
plt.show()
```

```python
import numpy as np
random_array = np.linspace(1,20,20)
print(random_array)
```

```python
# line plot
fig, ax = plt.subplots()
ax.plot(random_array, random_array*random_array);
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(random_array, random_array*random_array);
```

```python
# histogram plot
fig, ax = plt.subplots()
ax.hist(random_array, random_array*3);
```

```python
# bar plot
fig, ax = plt.subplots()
ax.set(title="sample title", xlabel="sample x label", ylabel="sample y label")
ax.bar(random_array, random_array*random_array);
```
